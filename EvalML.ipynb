```python
df =data.select_dtypes(include='object')
  
df=df.drop(['target_address'],axis=1)
for cols in df:
  dummies =pd.get_dummies(df[cols],drop_first=True)
  df = pd.concat([df,dummies],axis='columns')
data=pd.concat([data,df],axis='columns')
data=data.drop(['target_address','Unnamed: 0','event','cat_r','holiday','weather'],axis=1)
```

```python
import evalml
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X, y, problem_type='time series multiclass')
```

```python
X_train.shape                              
```

```python
from evalml.automl import AutoMLSearch

model = AutoMLSearch(X_train, y_train, problem_type="time series regression",
                      max_batches=1,problem_configuration={"gap": 0, "max_delay": 200,
                                             "forecast_horizon": 200, "time_index": "ds"})
model.search()
```

```python
model.rankings
```

```python
model.best_pipeline
```

```python
pl = model.best_pipeline
```

```python
pred = pl.predict_in_sample(X_test, y_test, objective=None, X_train=X_train, y_train=y_train)
```
